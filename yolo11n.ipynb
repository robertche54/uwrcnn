Train a nano model with the latest YOLO11 version.

```python
from ultralytics import YOLO

model = YOLO('yolo11n.pt')

results = model.train(seed=42, data='/content/drive/MyDrive/coco/configs/coco.yaml', epochs=100, imgsz=640)
```

Test our YOLO11n model.

```python
from ultralytics import YOLO

model = YOLO('runs/detect/train/weights/best.pt')

metrics = model.val(data='/content/drive/MyDrive/coco/configs/coco.yaml', split='test')
```
